# Visual Analysis to get inferences from the  data

```python
# Aim:
# Need to get the reasons for the Employees leaving the company
# suggestion to the HR so that they can retain thier employees
```

```python
df.columns
```

```python
# columns('satisfactoryLevel', 'lastEvaluation') are not much important as they are not helping us infer the reasons of leaving
```

```python
sns.countplot(data=df,x="numberOfProjects",hue="left")
plt.show()
```

```python
# inference- people who are assigned 2 projects and 7 projects are leaving most
# suggessions-I think projects from 7 projects people should be redistributed to people with 2 projects.
# (Assigning the 3-5)
```

```python
sns.histplot(data=df,x="avgMonthlyHours",hue="left")
plt.show()
```

```python
# people are leaving because of high consumption of monthly hours as people with avg monthly hours more than 275 and more all are leaving
# Monthly hours should be balanced
#giving extra incentives for extra working hours if employees are willing to do extra
```

```python
sns.countplot(data=df,x="timeSpent.company",hue="left")
plt.show()
```

```python
# inference= people leave the organisation after exp of 3 -5 years

# Once they complete 3-5 years, then they should be given supefor projectrvisory roles s with inexperienced ones
#working under them giving the role based incentives or increments
```

```python
sns.countplot(data=df,x="workAccident",hue="left")
plt.show()
```

```python

# inference=employees are not leaving just because of conflicts
# irrespective of work accident people are leaving
# suggession --nill
```

```python
sns.countplot(data=df,x="promotionInLast5years",hue="left")
plt.show()
```

```python
# People with no promotions in last 5 years are more likely to leave
# suggetions:HRs should provide time to time promotions to the emplyees
```

```python
sns.countplot(data=df,x="dept",hue="left")
plt.xticks(rotation=90)
plt.show()
```

```python
# inference= people of sales, technical, support are leaving the company most
# suggesions
# 1.give option to move other dep
# 2. incentives
# 3. for better team management team meetings to discuss whatever problem emplyees are facing
# 4.taking for team outings, lunch, for recreational activity and reduce stress level
```

```python
sns.countplot(data=df,x="salary",hue="left")
plt.show()
```

```python
# inferecences- employees with low/medium salary are leaving most
# suggestions- Regular sal hikes according to the experience and market value
# salary hikes + promotion + incentives + project changes
```
