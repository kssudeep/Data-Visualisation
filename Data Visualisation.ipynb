```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#linePlot
days = ['Monday','Tuesday','Wednesday','Thursday','Friday']
values = [29,19,5,11,24]
plt.plot(days,values)
plt.grid(True)
plt.show()
```

```python
plt.plot(days,values,color='red',marker='o',markerfacecolor='yellow',linestyle='dashed',markeredgecolor='black')
plt.grid(True)
plt.xlabel(days)
plt.ylabel(values)
plt.title("Line plot for sales and values")
plt.show()
```

```python
#BarGraph
Speed=[100,80,30,50,65]
Cars=['Alto','Urus','Verna','Mini','Jaguar']
plt.bar(Speed,Cars,width=10)
plt.ylabel("Cars name")
plt.xlabel("Speed of car")
plt.show()
```

```python
Speed = [100,80, 30, 50, 65]
Cars = ['Alto', 'Urus', 'Verna', 'Mini','Jaguar']

plt.bar(Speed, Cars, width=10, color=['red', 'yellow', 'green', 'orange','blue'], edgecolor='black')
plt.grid(True)
plt.ylabel("Cars name")
plt.xlabel("Speed of car")

length = len(Cars)

for i in range(length):
    plt.text(Speed[i], i, str(Speed[i]), ha='center')

plt.show()
```

```python
#Piecharts
Individuals =[10,20,30,10,10]
Agegroup=['0-18','18-25','25-50','50-75','Above 75']
plt.pie(Individuals,labels=Agegroup,autopct='%0.1f%%',colors=['Red','Pink','Blue','brown','purple'],shadow=True)
plt.show()
```

```python
plt.pie(Individuals, labels=Agegroup, autopct='%0.1f%%', colors=['Red', 'Pink', 'Blue', 'brown', 'purple'], shadow=True,explode=[0,0,0.1,0,0.8])
plt.show()
```

```python
Rollno = [5,7,8,10,12,15]
Marks=[34,45,55,23,78,66]
plt.scatter(Rollno,Marks,color='red')
plt.xlabel("Rollno")
plt.ylabel("Marks")
plt.show()
```

```python
#Scatter Plot
Rollno=[5,7,8,10,12,15]
Maths=[34,45,55,23,78,66]
Science=[12,89,45,75,56,99]
plt.scatter(Rollno,Maths,color='red',marker='*',label='math',s=200)
plt.scatter(Rollno,Science,color='Blue',marker='o',label='Science',s=200)
plt.xlabel("Rollno")
plt.ylabel("Marks")
plt.grid(True)
plt.legend(title='Marks',loc='lower right')
plt.show()
```

```python
#violin- plot
Groups=[45,56,75,34,89]
Data=[23,76,53,22,56,75]
newdata=[Groups,Data]
plt.boxplot(newdata,labels=['Groups','Data'])
plt.show()
```
